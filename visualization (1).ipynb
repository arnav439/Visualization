```python


#Create a dataset representing categories and their corresponding values. Compare different categories
#based on numerical values

categories = ['A', 'B', 'C', 'D']
values = {
    'Category': np.random.choice(categories, size=200),
    'Value': np.concatenate([
        np.random.normal(10,2, 50),
        np.random.normal(15, 3, 50),
        np.random.normal(20,4, 50),
        np.random.normal(25,5, 50),
    ])
}
# Convert to DataFrame
df = pd.DataFrame(values)

print(df.head())

plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df,palette='viridis')

plt.figure(figsize=(10, 6))
sns.violinplot(x='Category', y='Value', data=df,palette='viridis')
```

```python
#Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
#variable across different categories.
np.random.seed(42)

categories = ['A', 'B', 'C']
values = {
    'Category': np.random.choice(categories, size=300),
    'Value': np.concatenate([
        np.random.normal(loc=10, scale=2, size=100),
        np.random.normal(loc=15, scale=3, size=100),
        np.random.normal(loc=20, scale=4, size=100)
    ])
}


df = pd.DataFrame(values)


print(df.head())


plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Value', palette='Set2')
plt.title('Distribution of Numerical Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Value', hue='Category', multiple='stack', palette='Set2')
plt.title('Distribution of Numerical Values Across Categories ')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

```python
#Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
#heatmap.
n_samples = 200

x1= np.random.normal(0, 1, n_samples)
x2 = 2 * x1 + np.random.normal(0, 0.5, n_samples)
x3 = 0.5 * x1 + 0.3 * x2 + np.random.normal(0, 0.2, n_samples)
df = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3})

correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True,cmap = 'coolwarm')
```

```python
#Using the given x and y data, generate a 3D surface plot to visualize the function  Z = sin(sqrt(x.^2 + y.^2))
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale='Viridis')])

fig.update_layout(
    title='3D Surface Plot of Z = sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)
```

BOKEH ASSIGNMENT

```python
#Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x
import bokeh.io
from bokeh.plotting import figure, show ,output_file


x = np.linspace(0, 10,500)
y = np.sin(x)

output_file("sine_wave.html")

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="Sine Wave")
bokeh.io.output_notebook()
show(p)
```

```python
#Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
#markers based on the 'sizes' and 'colors' columns.
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral11

# Set a random seed for reproducibility
np.random.seed(42)


n = 200  # Number of points
x = np.random.random(n) * 10  # x values between 0 and 10
y = np.random.random(n) * 10  # y values between 0 and 10
sizes = np.random.random(n) * 30 + 5  # Marker sizes between 5 and 35
colors = np.random.choice(Spectral11, size=n)  # Random colors from the Spectral11 palette

source = ColumnDataSource(data={'x': x, 'y': y, 'sizes': sizes, 'colors': colors})
output_file("size and color.html")

# Create a Bokeh plot
p = figure(title="Bokeh Scatter Plot with Random Sizes and Colors",
           x_axis_label='X', y_axis_label='Y'
           )

p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6, legend_field='sizes', source=source)

output_notebook()
show(p)
```

```python
#Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
output_file("fruit_counts.html")
source = ColumnDataSource(data={'fruits': fruits, 'counts': counts})

p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruit', y_axis_label='Count')

p.vbar(x='fruits', top='counts', width=0.6, source=source, color="skyblue", legend_field="fruits")

output_notebook()
show(p)
```

```python
#Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

output_file("hist.html")
p = figure(title="Histogram of Random Data",
           x_axis_label='Value', y_axis_label='Frequency')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       color="skyblue", alpha=0.6, legend_label="Data Distribution")
show(p)
```
